```python
Base = automap_base()
Base.prepare(engine, reflect=True)
Base.classes['samples_metadata']
session = Session(bind=engine)
samples = session.query(Base.classes.samples_metadata).filter(Base.classes.samples_metadata.SAMPLEID=="941")
sm, = samples.all()
```

```python
Base = automap_base()
Base.prepare(engine, reflect=True)
sampleClass = Base.classes.samples
session = Session(bind=engine)
samples = session.query(sampleClass.otu_id,sampleClass.__dict__['BB_944']) \
.filter(sampleClass.__dict__['BB_944']>0) \
.order_by(sampleClass.otu_id.desc())
```
